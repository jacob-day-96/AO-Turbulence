```python
file = '/Users/jacob/Documents/Hawaii Research Project/master_files/master_file.fits'
hdul = fits.open(file)
data = hdul[1].data

dimm_values = data['dimm_value']
mass_values = data['mass_value']
ground_layer_values = data['seeing_gl']

MP_month = data['MP_month']
MP_year = data['MP_year']

CFHT_windspeed = data['CFHT_windspeed']
CFHT_wind_direction = data['CFHT_wind_direction']
CFHT_temp = data['CFHT_temp']
CFHT_humidity = data['CFHT_humidity']
CFHT_pressure = data['CFHT_pressure']
```

```python
# Wind Speed

# Wind speed values by month (1 = Jan, ..., 12 = Dec)
windspd_per_month = [CFHT_windspeed[MP_month == month] for month in range(1, 13)]

# Wind speed values by year (2009 to 2025)
windspd_per_year = [CFHT_windspeed[MP_year == year] for year in range(2009, 2026)]

#----------------------------------------------------------------------------------------------

# Wind Direction

# Wind direction values by month (1 = Jan, ..., 12 = Dec)
winddir_per_month = [CFHT_wind_direction[MP_month == month] for month in range(1, 13)]

# Wind direction values by year (2009 to 2025)
winddir_per_year = [CFHT_wind_direction[MP_year == year] for year in range(2009, 2026)]
#----------------------------------------------------------------------------------------------

# Wind Temperature

# Wind temperature values by month (1 = Jan, ..., 12 = Dec)
windtemp_per_month = [CFHT_temp[MP_month == month] for month in range(1, 13)]

# Wind temperature values by year (2009 to 2025)
windtemp_per_year = [CFHT_temp[MP_year == year] for year in range(2009, 2026)]
#----------------------------------------------------------------------------------------------

# Wind Humidity

# Wind humidity values by month (1 = Jan, ..., 12 = Dec)
windhum_per_month = [CFHT_humidity[MP_month == month] for month in range(1, 13)]

# Wind humidity values by year (2009 to 2025)
windhum_per_year = [CFHT_humidity[MP_year == year] for year in range(2009, 2026)]
#----------------------------------------------------------------------------------------------

# Wind Pressure

# Wind pressure values by month (1 = Jan, ..., 12 = Dec)
windpress_per_month = [CFHT_pressure[MP_month == month] for month in range(1, 13)]

# Wind pressure values by year (2009 to 2025)
windpress_per_year = [CFHT_pressure[MP_year == year] for year in range(2009, 2026)]
#----------------------------------------------------------------------------------------------

# Ground Layer Seeing Valyes

# Ground layer seeing values by month (1 = Jan, ..., 12 = Dec)
gl_per_month = [ground_layer_values[MP_month == month] for month in range(1, 13)]

# Ground layer seeing values by year (2009 to 2025)
gl_per_year = [ground_layer_values[MP_year == year] for year in range(2009, 2026)]
#----------------------------------------------------------------------------------------------

# DIMM Seeing Values

# DIMM seeing values by month (1 = Jan, ..., 12 = Dec)
dimm_per_month = [dimm_values[MP_month == month] for month in range(1, 13)]

# DIMM seeing values by year (2009 to 2025)
dimm_per_year = [dimm_values[MP_year == year] for year in range(2009, 2026)]
#----------------------------------------------------------------------------------------------

# Mass Values

# MASS seeing values by month (1 = Jan, ..., 12 = Dec)
mass_per_month = [mass_values[MP_month == month] for month in range(1, 13)]

# MASS seeing values by year (2009 to 2025)
mass_per_year = [mass_values[MP_year == year] for year in range(2009, 2026)]
```
