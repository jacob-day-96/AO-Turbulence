```python
# Data visualization notebook that uses the master dataset to compare
# monthly and yearly variation for each parameter using the `box_plot`
# function from `plot_functions.ipynb`.
# Example visualization shown below for the wind speed parameter.

file = '/path/to/master_data_file'
hdul = fits.open(file)
data = hdul[1].data

dimm_values = data['dimm_value']
mass_values = data['mass_value']
ground_layer_values = data['seeing_gl']

MP_month = data['MP_month']
MP_year = data['MP_year']

CFHT_windspeed = data['CFHT_windspeed']
CFHT_wind_direction = data['CFHT_wind_direction']
CFHT_temp = data['CFHT_temp']
CFHT_humidity = data['CFHT_humidity']
CFHT_pressure = data['CFHT_pressure']
```

```python
# Wind Speed

# Wind speed values by month (1 = Jan, ..., 12 = Dec)
windspd_per_month = [CFHT_windspeed[MP_month == month] for month in range(1, 13)]

# Wind speed values by year (2009 to 2025)
windspd_per_year = [CFHT_windspeed[MP_year == year] for year in range(2009, 2026)]
```

### Wind Speed
