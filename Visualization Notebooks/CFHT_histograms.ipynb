```python
# Data visualization notebook that uses the master dataset to generate
# histograms and CDFs for selected parameters using the `doane_hist`
# function from `plot_functions.ipynb`.
# Visualizations use Doane’s binning rule to highlight the overall
# distribution and statistical characteristics of the data.

file = 'path/to/master_file'
hdul = fits.open(file)
data = hdul[1].data 

wind_speed = data['wind_speed']
wind_dir = data['wind_dir']
temp = data['temp']
humidity = data['humidity']
pressure = data['pressure']

ground_layer = data['ground_layer']
mass_val = data['mass_val']
```

## Doane's Rule

$$
k = 1 + \log_{2}(n) + \log_{2}(1 + \frac{|\text{skewness}|}{\sigma_{g1}})
$$
$$
k = \text{The number of bins}
$$
$$
n = \text{The number of data points}
$$
$$
\text{skewness = sample skewness of the data}
$$
$$
\sigma_{g1} = \sqrt{\frac{6(n-2)}{(n+1)(n+3)}} \text{(standard error of skewness)}
$$

### CFHT Wind Speed

```python
doane_hist(wind_speed, 'Wind Speed (m/s)')
```

```python
plt.figure(figsize=(8,4))
plt.hist(wind_speed, bins = 40)
plt.axvline(np.median(wind_speed), color='red')
plt.ylabel('Frequency')
plt.xlabel('Wind Speed (m/s)')
plt.show()

print('The max value is: ', np.max(wind_speed))
print('The min value is: ', np.min(wind_speed))
print('The median value is: ', np.median(wind_speed))
print('The std value is: ', round(np.std(wind_speed),3))
```
